```python
import numpy as np
import os
import pandas as pd
from PIL import Image
```

```python
PARENT_DIR = '/media/johnkoo/data/stanford-dogs-dataset'
IMAGE_SHAPE = (128, 128)
```

```python
os.chdir(PARENT_DIR)
meta_df = pd.read_csv('meta.csv')
meta_df
```

```python
os.chdir(PARENT_DIR)
os.chdir('Images')

dogs = np.unique(meta_df.label)

val_images = []
val_labels = []
test_images = []
test_labels = []

for subset in ['val', 'test']:
    sub_df = meta_df[meta_df.dataset == subset]
    
    for dog in dogs:
        dog_df = sub_df[sub_df.label == dog]
        os.chdir(dog)
        
        for i, file in enumerate(dog_df.filename):
            image = np.array(Image.open(file).resize(IMAGE_SHAPE))[:, :, 0:3]
            
            if subset == 'val':
                val_images.append(image)
                val_labels.append(dog)
            else:
                test_images.append(image)
                test_labels.append(dog)
        os.chdir('..')
```

```python
print(len(val_images))
print(len(val_labels))
print(len(test_images))
print(len(test_labels))
```

```python
val_images = np.stack(val_images)
val_labels = np.array(val_labels)
test_images = np.stack(test_images)
test_labels = np.array(test_labels)
```

```python
os.chdir(PARENT_DIR)

if 'val' not in os.listdir():
    os.mkdir('val')
os.chdir('val')
np.save('./val-images.npy', val_images)
np.save('./val-labels.npy', val_labels)

os.chdir('..')

if 'test' not in os.listdir():
    os.mkdir('test')
os.chdir('test')
np.save('./test-images.npy', test_images)
np.save('./test-labels.npy', test_labels)
```
