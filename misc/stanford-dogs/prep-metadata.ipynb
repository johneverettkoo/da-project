```python
import xmltodict
import os
import pandas as pd

PARENT_DIR = '/media/johnkoo/data/stanford-dogs-dataset'
TRAIN_SIZE = 100
VAL_SIZE = 25
```

```python
os.chdir(PARENT_DIR)
os.chdir('Annotation')
unique_labels = os.listdir()
```

```python
os.chdir(PARENT_DIR)
os.chdir('Annotation')

labels = []
filenames = []
centroid_1 = []
centroid_2 = []
coverage = []
dataset = []

for dog in unique_labels:
    os.chdir(dog)
    files = os.listdir()
    
    i = 0
    
    for file in files:
        with open(file) as f:
            meta = xmltodict.parse(f.read())['annotation']
        
        if isinstance(meta['object'], dict):
            total_size = float(meta['size']['width']) * float(meta['size']['height'])
            bbox = meta['object']['bndbox']
            bbox_size = (float(bbox['xmax']) - float(bbox['xmin'])) * (float(bbox['ymax']) - float(bbox['ymin']))
            cover = bbox_size / total_size

            c1 = (float(bbox['ymax']) + float(bbox['ymin'])) / 2
            c2 = (float(bbox['xmax']) + float(bbox['xmin'])) / 2

            labels.append(dog)
            filenames.append(file + '.jpg')
            centroid_1.append(c1)
            centroid_2.append(c2)
            coverage.append(cover)

            if i < TRAIN_SIZE:
                dataset.append('train')
            elif i < TRAIN_SIZE + VAL_SIZE:
                dataset.append('val')
            else:
                dataset.append('test')

            i += 1
    os.chdir('..')
```

```python
print(len(labels))
print(len(filenames))
print(len(centroid_1))
print(len(centroid_2))
print(len(coverage))
print(len(dataset))
```

```python
meta_df = pd.DataFrame({'label': labels, 
                        'filename': filenames, 
                        'centroid_1': centroid_1, 
                        'centroid_2': centroid_2, 
                        'coverage': coverage, 
                        'dataset': dataset})
```
